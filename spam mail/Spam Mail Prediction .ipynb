# SPAM MAIL DETECTION

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
mail_data=pd.read_csv('spam.csv',encoding='latin-1')
```

```python
mail_data.head()
```

```python
mail_data.isnull().sum()
```

```python
mail_data.shape
```

```python
for column in mail_data.columns:
    print(f"Column: {column}")
    print(len(mail_data[column].unique()))
```

```python
# replace the null values with a null string
new_mail_data = mail_data.where((pd.notnull(mail_data)), '')
```

```python
new_mail_data.head()
```

### Label Encoding

```python
new_mail_data.loc[new_mail_data['v1'] == 'spam', 'v1'] = 0
new_mail_data.loc[new_mail_data['v1'] == 'ham', 'v1'] = 1
```

spam - 0<br>
ham - 1

```python
X = new_mail_data['v2']
Y = new_mail_data['v1']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
vectorizer = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)
X_train_features = vectorizer.fit_transform(X_train)
X_test_features = vectorizer.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_train_features)
```

```python
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
model = LogisticRegression()
model.fit(X_train_features, Y_train)
```

```python
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print('Accuracy on test data : ', accuracy_on_test_data)
```
